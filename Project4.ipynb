```python
!pip install folium
import folium
import requests
import pandas

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

arrest_table
```

```python
arrest_table.pickup_datetime = pandas.to_datetime(arrest_table.arrestDate, format='%m/%d/%Y')
arrest_table['month'] = arrest_table.pickup_datetime.apply(lambda x: x.month)
arrest_table['day'] = arrest_table.pickup_datetime.apply(lambda x: x.day)
arrest_table
```

Above I changed the datetime to add new rows for months and days. After this I created the table below which only includes the month of January. This was to have about 1/12th the data so my data doesn't clutter.

```python
heat_map = arrest_table[arrest_table['month']==1]
heat_map
```

Below is a heatmap of the "intensity" of crime in a particular area during the month of January.To do this, I counted the amount of times a crime happened at a location. After summing all crimes at each longitude and latitude. I displayed it on the map below.

A heatmap has a radius where it blends with other locations. Every overlap from one circle to another will increase the intensity of the color. Blue means that there is relatively few crimes in that areas. The more redder areas represent areas where there is ahigher amount of crime

As shown most crime happens in downtown baltimore where there are a few hotspots where crime is slightly higher. There Is also a large crime spot Northwest of the downtown. Overall there seems to be areas where crimes happen more frequently then other areas of baltimore. This could be because that area is more dense or simply has a higher population. Overall, the data can't accuratly display "crimes per capita" but it can show where crimes are occuring in the city.

```python
from folium.plugins import HeatMap
heat_map['count'] = 1
#df = heat_map[['lat', 'long', 'count']].groupby(['lat', 'long']).sum().reset_index()
d={}
for row_index,row in heat_map.iterrows():
    if d.__contains__((row["lat"],row["long"])):
        d[(row["lat"],row["long"])] = d[(row["lat"],row["long"])] + 1
    else:
         d[(row["lat"],row["long"])] = 1
g=[[]]
for key,value in d.items():
    
    g.append([key[0],key[1],value])
g.pop(0)


HeatMap(data=g, radius=6, max_zoom=13).add_to(map_osm)
map_osm
```

Below is a map of the sex of the crime in balitmore during the month of January. For every male crime I put a blue dot. For every Female crime, I put a yellow dot. As shown most crime are committed by males. There seems to be no correlation between sex and location. 

```python
for row_index,row in heat_map.iterrows():
    if row["sex"] == "M":
        folium.Circle(
        radius=10,
        location=[row["lat"], row["long"]],
        color='blue',
        fill=False,
        ).add_to(map_osm2)
    else:
        folium.Circle(
        radius=10,
        location=[row["lat"], row["long"]],
        color='yellow',
        fill=False,
        ).add_to(map_osm2)
map_osm2
```
